```python
from googleapiclient.discovery import build
import os
from dotenv import load_dotenv

# Load variables from .env file
load_dotenv()
```

```python
def get_channel_videos(api_key, channel_id):
    youtube = build('youtube', 'v3', developerKey=api_key)
    
    # 1. Get the 'uploads' playlist ID for the channel
    ch_request = youtube.channels().list(
        part='contentDetails',
        id=channel_id
    )
    ch_response = ch_request.execute()
    uploads_playlist_id = ch_response['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    # 2. Extract video titles from that playlist
    video_data = []
    next_page_token = None
    
    while True:
        pl_request = youtube.playlistItems().list(
            part='snippet',
            playlistId=uploads_playlist_id,
            maxResults=50,
            pageToken=next_page_token
        )
        pl_response = pl_request.execute()
        
        for item in pl_response['items']:
            title = item['snippet']['title']
            video_id = item['snippet']['resourceId']['videoId']
            url = f"https://www.youtube.com/watch?v={video_id}"
            
            video_data.append({
                'title': title,
                'url': url
            })
            
        next_page_token = pl_response.get('nextPageToken')
        if not next_page_token:
            break
            
    return video_data

# Usage
MY_KEY = os.getenv('YOUTUBE_API_KEY')
CHANNEL_ID = 'UCAHr-sT0AjrD3sBwr1eRUNg' # Mark Meldrum Channel ID
videos = get_channel_videos(MY_KEY, CHANNEL_ID)
print(f"Found {len(videos)} videos.")
```
